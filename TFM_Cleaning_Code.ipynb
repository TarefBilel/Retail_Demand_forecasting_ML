## First part

## 1. Dataset link

https://www.kaggle.com/datasets/asahu40/walmart-data-analysis-and-forcasting


## 2. Preprocessing the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_1=pd.read_csv("/content/drive/MyDrive/TFM/Walmart Data Analysis and Forcasting.csv",parse_dates=["Date"])
data_1
```

```python
data_1.info()
```

```python
data_1["Date"]=pd.to_datetime(data_1["Date"],format='%d-%m-%Y')
```

```python
data_1.info()
```

```python
df_agg=data_1.groupby("Date",as_index=False).agg({
    "Weekly_Sales":"sum",

})[["Date","Weekly_Sales"]]
print(df_agg)
```

```python
df_agg
```

```python
df_agg.to_csv(path_or_buf="/content/drive/MyDrive/TFM/Walmart_Cleaned.csv")
```

## 2. Preprocessing the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_3=pd.read_csv("/content/drive/MyDrive/TFM/Supplement_Sales_Weekly_Expanded.csv",parse_dates=["Date"])
df_3.head(10)
```

```python
df_3["Date"]=pd.to_datetime(df_3["Date"],format='%d-%m-%Y')
```

```python
df_3.info()
```

```python
df_3
```

```python
df_3_agg=df_3.groupby("Date",as_index=False).agg({
    "Revenue":"sum",

})[["Date","Revenue"]]
print(df_3_agg)
```

```python
df_3_agg.to_csv(path_or_buf="/content/drive/MyDrive/TFM/Supplement_Sales_Cleaned.csv")
```
